```python
# populate OpenSearch index using python script through FluentD
import socket, msgpack, time
import httpx
import asyncio
import json

offer = {
    "uuid": "123e4567-e89b-12d3-a456-426614174000",
    "name": "Sample Product",
    "description": "This is a sample product description.",
    "price": 19.99,
    "currency": "USD",''
    "seller_uuid": "uuid-seller-0001"
    }

async def send_to_fluentd_async():
    """Send data to Fluentd using async httpx client"""
    
    async with httpx.AsyncClient(timeout=10.0) as client:
        try:
            # TODO: add advanced step to get more info from another service before sending to Fluentd
            # Send to Fluentd via HTTP
            fluentd_response = await client.post(
                "http://localhost:9880/products",
                headers={
                    "Content-Type": "application/json",
                    "User-Agent": "jupyter-notebook/httpx"
                },
                json=offer,
                timeout=5.0
            )
            
            print(f"Fluentd HTTP Response: {fluentd_response.status_code}")
            print(f"Response headers: {dict(fluentd_response.headers)}")
            
            if fluentd_response.text:
                print(f"Response body: {fluentd_response.text}")
                
        except httpx.TimeoutException:
            print("Request timed out")
        except httpx.ConnectError:
            print("Connection error - check if Fluentd is running on port 9880")  
        except httpx.HTTPStatusError as e:
            print(f"HTTP error: {e.response.status_code} - {e.response.text}")
        except Exception as e:
            print(f"Unexpected error: {e}")

# Run the async function
await send_to_fluentd_async()
```
